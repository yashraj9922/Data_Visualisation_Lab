```python
## Bonus: How does the weight of the cars (wt) correlate with miles per gallon (mpg)?
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='wt', y='mpg', data=mtcars, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Correlation between Weight and Miles per Gallon')
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Miles per Gallon')
plt.show()
```

```python
## Bonus: Distribution of Displacement (disp) across Different Cylinder Types (cyl)
# Boxplot of displacement across different cylinder types
plt.figure(figsize=(8, 6))
sns.boxplot(x='cyl', y='disp', data=mtcars)
plt.title('Distribution of Displacement by Cylinder Type')
plt.xlabel('Number of Cylinders')
plt.ylabel('Displacement (cu.in.)')
plt.show()
```

```python
##Bouns:  Effect of Number of Carburetors (carb) on Horsepower (hp)
```

```python
# Boxplot of horsepower across different numbers of carburetors
plt.figure(figsize=(8, 6))
sns.boxplot(x='carb', y='hp', data=mtcars)
plt.title('Effect of Number of Carburetors on Horsepower')
plt.xlabel('Number of Carburetors')
plt.ylabel('Horsepower')
plt.show()
```

Weight vs. Miles per Gallon: The scatter plot with a regression line helps visualize the negative correlation between a car's weight and its fuel efficiency.

Displacement by Cylinder Type: The boxplot shows how displacement varies across different cylinder types, helping to identify any trends or outliers within each group.

Carburetors vs. Horsepower: The boxplot illustrates how the number of carburetors impacts a car's horsepower, showing the distribution of horsepower values for each carburetor count.

```python
import pandas as pd

# Load the Datasaurus Dozen dataset
datasaurus = pd.read_csv('DatasaurusDozen.tsv', sep='\t')

# Preview the dataset
print(datasaurus.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for the plots
sns.set(style="whitegrid")

# Create a scatter plot for each dataset
g = sns.FacetGrid(datasaurus, col="dataset", col_wrap=4, height=4)
g.map(sns.scatterplot, "x", "y", s=50, color="b")

# Adjust the layout for better visibility
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Datasaurus Dozen: Visualization of Different Datasets')

plt.show()
```

FacetGrid: This is used to create a grid of scatter plots, one for each subset of the data defined by the dataset column. Each scatter plot visualizes the x and y coordinates.

Variation in Plots: Despite the datasets having similar summary statistics, the scatter plots will reveal very different shapes, patterns, and distributions, such as dinosaurs, stars, and other shapes. This stark visual difference underscores the importance of plotting data rather than relying solely on numerical summaries.
